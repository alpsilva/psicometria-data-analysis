# Etnografia

Distribuição dos campos:
- Faixa Etária
- Gênero
- Orientação Sexual
- Etnia

```python
def category_plot(df: pandas.DataFrame, fig_name: str, label: str, order: list = None, label_as_title: bool = False):
    config = {
        'data': df,
        'x': label,
        'stat': "percent",
        'legend': "full"
    }

    if order is not None:
        config['order'] = order

    plt.figure(fig_name)
    fig = sns.countplot(**config)
    
    if label_as_title:
        fig.set(title=label)

    plt.tight_layout()
    plt.show()
```

```python
age_order = [
    "Entre 18 e 24",
    "Entre 25 e 39",
    "Entre 40 e 65",
    "Acima de 65"
]

category_plot(df, "a", "Faixa Etária" , age_order, True)
```

```python
category_plot(df, "b", "Gênero", label_as_title=True)
```

```python
category_plot(df, "c", "Orientação Sexual", label_as_title=True)
```

```python
category_plot(df, "d", "Etnia", label_as_title=True)
```

# Distribuição das perguntas
Distribuição das respostas para as seguintes perguntas, em ordem:


```python
questions = [
    'Preocupo-me com a segurança das pessoas que não têm a mesma visão política que eu.',
    'Fico comovida(o) quando alguém é tratado injustamente por ter posicionamento político divergente dos demais.',
    'Sinto compaixão por pessoas com visões políticas diferentes da minha.',
    'Sinto tristeza se coisas ruins acontecem com pessoas de visões políticas diferentes das minhas.',
    'Sinto-me feliz ao ver pessoas com opiniões políticas diferentes das minhas prosperarem.',
    'Permaneço receptivo às posições políticas alheias.',
    'Analiso todas as diferentes perspectivas políticas disponíveis.',
    'Entendo a posição política dos outros, mesmo quando diferente da minha.',
    'Imagino como as pessoas sentem-se quando eu critico suas visões políticas.',
    'Eu me coloco no lugar das outras pessoas para entender suas perspectivas políticas.',
    'Eu me relaciono bem com pessoas que possuem visões políticas diferentes das minhas.',
    'Dialogo com pessoas que têm opiniões políticas diferentes das minhas.',
    'Resolvo minhas diferenças com pessoas com visões políticas divergentes das minhas.',
    'Nos embates políticos estou atenta(o) a todas as opiniões.',
    'Trabalho em projetos políticos de redução da polarização política.'
]
```

```python
answear_order = [1, 2, 3, 4, 5]

for i, question in enumerate(questions):
    question_num = i + 1
    print(f"Pergunta {question_num}: {question}")
    category_plot(df, str(i), question, order=answear_order, label_as_title=True)
```
